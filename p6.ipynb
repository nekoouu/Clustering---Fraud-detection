```python
def elbow_for_kmodes(data,a,b):
    # Elbow curve to find optimal K
    from kmodes.kmodes import KModes
    cost = []
    K = range(a,b)
    for num_clusters in list(K):
        kmode = KModes(n_clusters=num_clusters, init = "random", n_init = 5, verbose=1)
        kmode.fit_predict(data)
        cost.append(kmode.cost_)

    plt.plot(K, cost, 'bx-')
    plt.xlabel('No. of clusters')
    plt.ylabel('Cost')
    plt.title('Elbow Method For Optimal k')
    plt.show()

#K-modes
def k_modes(data, k):
    import numpy as np
    from kmodes.kmodes import KModes
    import matplotlib.pyplot as plt
    kmode = KModes(n_clusters=k, init = "random", n_init = 5, verbose=1)
    clusters = kmode.fit_predict(data)
    return clusters
```

```python
# Agglomerative Clustering
def agglo(data, k ):
    from sklearn.cluster import AgglomerativeClustering
    AC = AgglomerativeClustering(n_clusters=k)
    # fit model and predict clusters
    yhat_AC = AC.fit_predict(data)
    result_data = data
    result_data["Clusters"] = yhat_AC

    return result_data
```

```python
result_data = agglo(PCA_df,2)

from sklearn.metrics import silhouette_score
s_a = silhouette_score(PCA_df, result_data["Clusters"])

from sklearn.metrics import accuracy_score
a_a = accuracy_score(y, result_data["Clusters"])

#Plotting the clusters
fig = plt.figure(figsize=(10,8))
ax = plt.subplot(111, projection='3d', label="bla")
ax.scatter(z_1, z_2, z_3, s=40, c=result_data["Clusters"], marker='o', cmap = "plasma" )
ax.set_title("The Plot Of The Clusters")
plt.show()
```

```python
#DBSCAN
def dbscan(X, e, n):
    from sklearn.cluster import DBSCAN

    db = DBSCAN(eps=e, min_samples=n).fit(X)
    labels = db.labels_
    result_data = X
    result_data["Clusters"] =  labels

    n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise_ = list(labels).count(-1)

    return result_data, n_clusters_

#Gridsearch
def gridsearch_4_dbscan(X, param_grid):
    from sklearn.cluster import DBSCAN
    from sklearn.model_selection import GridSearchCV

    dbscan = DBSCAN()
    grid_search = GridSearchCV(dbscan, param_grid,  scoring='accuracy')
    grid_search.fit(X)

    #best_eps = grid_search.best_params_['eps']
    #best_min_samples = grid_search.best_params_['min_samples']

    return grid_search.best_estimator_.get_params()
```

```python
import numpy as np
epsilon = np.linspace(0.01,1,num = 10)
min_sample = np.arange(2,12,step = 5)

import itertools
combinations = list(itertools.product(epsilon,min_sample))
N = len(combinations)

param_dict = {}
param_dict["eps"] = []
param_dict["min_samples"] = []

for i in range(0,N):
    param_dict["eps"].append(combinations[i][0])
    param_dict["min_samples"].append(combinations[i][1])

param_dict["eps"]= list(set(param_dict["eps"]))
param_dict["min_samples"]= list(set(param_dict["min_samples"]))
```
